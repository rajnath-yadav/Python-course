## Desafio

**6)** O sistema de amortização por [Tabela Price](https://pt.wikipedia.org/wiki/Tabela_Price?oldformat=true) é um método utilizado para amortização de empréstimos e dívidas, que apresenta como principal característica, prestações/parcelas constantes.

Apesar do valor da parcela ser constante, a cada período, uma parte dessa parcela é dedicada para os juros da operação, enquanto outra parte é dedicada ao abatimento do valor inicial (amortização do principal). Os juros são calculados com base no principal devido, então a tendência natural é que o valor da parcela dedicado aos juros seja grande no início, e vai caindo conforme o valor devido for sendo amortizado.

Veja abaixo um exemplo de uma Tabela Price, de um empréstimo de R$200 mil tomados à taxa de juros de 0.6% a.m com um financiamento de 30 anos (360 meses). A coluna Saldo Inical mostra o quanto do valor principal era devido naquele determinado mês antes do pagamento, e a coluna Saldo Final mostra o saldo devedor após a amortização.

Observe que os valores de juros e amortização são lançados como negativos pois são despesas para o devedor.

Acima, o devedor tem uma despesa mensal fixa de 1357 reais (juros + amortizações). Note que porção dedicada ao pagamento de juros é bem alta nos primeiros meses, e bem baixa nos últimos, dando espaço para o crescimento da porção de amortizações ao longo dos meses.

A proposta do desafio é a implementação de uma classe chamada `price` que entre outras funcionalidades, irá gerar um DataFrame similar ao apresentado acima.

### Inicialização
A função `__init__` receberá os parâmetros `vp, r` e  `n`, respectivamente valor presente (dívida inicial contratada), taxa de juros (na forma decimal), e quantidade de períodos. Aqui as unidades de `r` e `n` serão passadas de forma consistente, por exemplo, se a taxa for mensal, `n` deverá ser passado em meses.

Além disso você deve criar um novo atributo `parcela` que irá armezanar o valor da parcela a ser paga a cada período. O cálculo segundo é feito pela fórmula abaixo:

\begin{equation}
\dfrac{VP\cdot r}{1-(1+r)^{-n}}
\end{equation}

Veja um exemplo de intância da classe do exemplo dado acima:

>>>> minha_divida.vp
>200000
>>>> minha_divida.r
>0.006
>>>> minha_divida.n
>360
>>>> minha_divida.parcela
>1357.5764189349288

### Saldo Devedor

### Arrays de Juros e Amortizações

Complete a implementação dos métodos `juros` e `amortizacao`. A primeira deve retornar uma array numpy de tamanho `n` onde a i-ésima entrada contém o pagamento de juros no período i. A segunda retorna uma array de mesmo tamanho, porém a i-ésima entrada conterá o valor de amortização pago no i-ésimo período. 

**ATENÇÃO:** Os valores internos das arrays deverão ser negativos, pois representam um gasto para o devedor, essa é a convenção utilizada no mercado!

Dado um período t, o módulo do pagamento de juros nada mais é do que uma multiplicação da taxa r, pelo saldo devedor principal do período t-1.

A amortização de um determinado período i, pode ser calculada de duas formas:

    1 - Lembrando que o módulo de (juro + amortização) é igual à parcela fixa.

    2 - Pela fórmula:

\begin{equation}
A_t = A_1 \cdot (1+r)^{t-1}
\end{equation}

onde A<sub>t</sub> é a porção de amortização a ser paga no período t, e A<sub>1</sub> é a amortização no período 1

Veja a execução dessas funções no exemplo dos R$200 mil:

>```python
>>>> minha_divida = price(200000,.006,360)
>>>> minha_divida.juros()
>array([-1200., -1199.05454149, -1198.10341022, 
>        ..., -24.14604523,   -16.14546299,    -8.09687725])
>>>> minha_divida.amortizacao()
>array([ -157.57641893,  -158.52187745,  -159.47300871, 
>        ..., -1333.43037371, -1341.43095595, -1349.47954168])
>```
