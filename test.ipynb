```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = {"col1": ["one","three","three","three","two","two","four","four","four","four"],
     "col2": ["A","B","A","B","C","A","B","C","D","A"],
     "col3": [1,5,2,3,4,5,6,7,5,6]}
```

```python
d
```

```python
df = pd.DataFrame(d)
df.head(10)
```

```python
df.col1.value_counts()
```

```python
df.col2.value_counts()
```

```python
df.iloc[:,0].value_counts().plot(kind="bar", stacked=True)
```

```python
labels = df.columns[0]
width = 0.25 

fig, ax = plt.subplots(figsize=(4,10))

b = 0
for stack in range(0,len(df.iloc[:,0].value_counts().values)):
    p1 = ax.bar(labels, 
                df.iloc[:,0].value_counts().values[stack], 
                width, 
                label=df.iloc[:,0].value_counts().index[stack],
                bottom=b,
               )
    b += df.iloc[:,0].value_counts().values[stack]
    ax.bar_label(p1, label_type='center')

ax.set_ylabel('Counts')
ax.set_title('My Title')
ax.legend()

plt.show()
```
